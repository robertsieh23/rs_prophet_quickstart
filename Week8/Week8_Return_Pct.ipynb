```python
import pandas as pd
import os
```

```python
frames = {}
path = 'github_data/'
for filename in os.listdir(path):
    coin = filename.split('.')[0]
    frames[coin] = pd.read_csv(path + filename)
```

```python
for coin, df in frames.items():
    frames[coin]['return_pcnt'] = (df.close - df.open) / df.close
    frames[coin]['volatility_pcnt'] = (df.high - df.low) / ((df.high + df.low) / 2)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
for coin, df in frames.items():
    
    fig = plt.figure(figsize=(15,5))
    fig.suptitle(coin, size=20)
    ax = plt.gca()
    plt.plot(df.index, df.return_pcnt, color='black')
    for ix, label in enumerate(ax.xaxis.get_ticklabels()):
        label.set_visible(True) if ix % 80 == 0 else label.set_visible(False)
    ax.set_ylabel('USD', size=16)
    ax.set_title('Daily Return Percentage', size=16)
    plt.show()
```

```python
print("correlations")
for coin, df in frames.items():
    print(coin)
    print(df.corr()[['return_pcnt', 'volatility_pcnt']].loc[['return_pcnt']])
```
