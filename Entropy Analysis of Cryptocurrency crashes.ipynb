Abstract

This project investigates the relationship between exchange rate entropy and exchange prices for cryptocurrencies. The main focus here is during crisis periods, when the exchange rate rapidly decreases. The analysis shows entropy spikes associated with cryptocurrencies crises in both crypto to fiat and crypto to bitcoin markets.

Introduction

There's been research in the FX market relating increased amounts of exchange rate entropy with Financial crises. Since cryptocurrencies are a type of currency, we can check to see if there is similar behavior in crypto to fiat markets and crypto to BTC markets.

Methodology

We use log return data at 4 hour intervals collected from Poloniex. To convert time series price data to entropy measurements, we first sigma encode the log return values into std deviation width buckets. Then we use rolling windows of 100 days to calculate the entropy using the plugin method. The entropy values are then graphed alongside logarithmic price data to investigate similarities

```python
import pandas as pd
import numpy as np
import scipy as sp
import requests
from datetime import timedelta
import matplotlib.pyplot as plt
import time
import math
```

```python
POLONIEX_OHLCV_BASEURL = 'https://poloniex.com/public?command=returnChartData&currencyPair='

"""
https://poloniex.com/public?command=returnChartData&currencyPair=BTC_POT&start=1435699200&end=9999999999&period=14400

"""

def get_ohlcv_poloniex(pair='USDT_BTC', start=1435699200, end=9999999999, period=900):
    """
    returns ohlcv data for poloniex as pandas dataframe
    convert to unix timestamp using https://coderstoolbox.net/unixtimestamp/
    :param pair: str pair on poloniex
    :param start: int unix timestamp of beginning time
    :param end: int unix timestamp of ending time
    :param period: int candle width in seconds
    :return: pandas df of ohlcv data from poloniex for specified pair, times, and period
    """
    query = POLONIEX_OHLCV_BASEURL + pair + '&start=' + str(start) + '&end=' + str(end) + '&period=' + str(period)
    resp = requests.get(query)

    if resp.status_code != 200:
        raise requests.ApiError('GET /tasks/ {}'.format(resp.status_code))

    return pd.DataFrame(resp.json())
```

```python
def plugIn(msg, w):
    # Compute plug-in (ML) entropy rate
    pmf = pmf1(msg, w)
    out = - sum([pmf[i] * np.log2(pmf[i]) for i in pmf]) / w
    return out, pmf

def pmf1(msg, w):
    # Compute the prob mass function for a 1D discrete RV
    # len(msg)-w occurances
    lib = {}
    if not isinstance(msg, str): msg = ''.join(map(str, msg))
    for i in range(w, len(msg)):
        msg_ = msg[i-w: i]
        if msg_ not in lib: 
            lib[msg_] = [i-w]
        else: 
            lib[msg_] = lib[msg_] + [i-w]
    pmf = float(len(msg) - w)
    pmf = {i: len(lib[i])/pmf for i in lib}
    return pmf
```

```python
def graph_entropy(pair):
    timeSeries = pd.DataFrame()
    df = get_ohlcv_poloniex(pair=pair, start=0, end=9999999999, period=14400)
    df['log_ret'] = np.log(df['weightedAverage']) - np.log(df['weightedAverage'].shift(1))
    df = df[df['log_ret'] != 0]
    df = df[df['log_ret'] != np.inf]
    df = df[df['log_ret'].notnull()]
    df['log_price'] = np.log(df['weightedAverage'])
    
    #sigma encoding
    edges = []
    std = df['log_ret'].std()
    edge = df['log_ret'].min()
    ceiling = df['log_ret'].max()
    while (edge < ceiling):
        edges.append(edge)
        edge += std
    edges.append(edge)
    df['sigma'] = pd.cut(df['log_ret'], edges, right=False, labels=False)
    df['entropy'] = df['sigma'].rolling(400).apply(lambda x: plugIn(x, 10)[0]) #100 day windows
    
    df['time'] = pd.to_datetime(df['date'], unit='s')
    df = df.set_index('time')
    
    plt.figure(figsize=(12,5))
    plt.xlabel(pair)

    ax1 = df.log_price.plot( color='blue', grid=True, label='log price')
    ax2 = df.entropy.plot(color='red', grid=True, secondary_y=True, label='entropy')

    h1, l1 = ax1.get_legend_handles_labels()
    h2, l2 = ax2.get_legend_handles_labels()


    plt.legend(h1+h2, l1+l2, loc=2)
    plt.show()
```

Discussion
We see entropy spikes in periods associated with large selloffs throughout the various exchange rates (e.g, 2015, 2018 selloffs). These are similar to financial crises given the frenzied large-scale sells. As a bear market continues to decrease, entropy falls as well. The entropy spikes are more pronounced in BTC denominated markets, which makes sense given that altcoins exhibit more volatile behavior than Bitcoin, and have a higher composition of short-term traders. The fiat markets also experience a buildup of entropy during rapid paced bull markets.
