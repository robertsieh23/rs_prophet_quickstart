```python
import pandas as pd
import os
```

```python
frames = {}
path = 'github_data/'
for filename in os.listdir(path):
    coin = filename.split('.')[0]
    frames[coin] = pd.read_csv(path + filename)
```

To rank for healthy developer activity, we're going to use sparsity of the fundemental features. Healthy projects ideally have developers and stargazers conducting activity every day. Thus we'll rank by sparsity (low to high) for each of these features and then do a rank consisting a sum of the normalized sparsities.

```python
feature_sparsities = {}
for feature in ['pr_count', 'issues_count', 'commit_add_sum', 'commit_del_sum', 'commit_count', 'star_count']:
    feature_sparsities[feature] = {}
    for coin, df in frames.items():
        feature_sparsities[feature][coin] = sum(df[feature] == 0)/len(df)
    print (feature)
    print(sorted(feature_sparsities[feature].items(), key=lambda x: x[1]))
    print()
#feature_sparsities
```

```python
#from https://stackoverflow.com/questions/16417916/normalizing-dictionary-values
def normalize(d, target=1.0):
   raw = sum(d.values())
   factor = target/raw
   return {key:value*factor for key,value in d.items()}
```

```python
for key, val in feature_sparsities.items():
    feature_sparsities[key] = normalize(val)
```

```python
normalized_sums = {}
for coin, y in next(iter(feature_sparsities.values())).items():
    sum = 0
    for key, val in feature_sparsities.items(): 
        sum += val[coin]
    normalized_sums[coin] = sum
print("Normalized Sum Ranking")
sorted(normalized_sums.items(), key=lambda x: x[1])
```
